# Spliter

```python
import fitz  # PyMuPDF
from langchain_core.documents import Document
from langchain_text_splitters import RecursiveCharacterTextSplitter

class PersianPDFSplitter:
    def __init__(
        self,
        chunk_size=8000,
        chunk_overlap=20,
        length_function=len,
        fall_back_splitter=RecursiveCharacterTextSplitter,
        fall_back_splitter_kwargs={},
    ):
        self.__len = length_function
        self.chunk_size = chunk_size
        self.chunk_overlap = chunk_overlap
        self.fall_back_splitter = fall_back_splitter(
            chunk_size=chunk_size,
            chunk_overlap=chunk_overlap,
            **fall_back_splitter_kwargs,
        )

    def create_documents(self, pdf_addr):
        doc = fitz.open(pdf_addr)
        text = ""
        # Extract all text from PDF
        for page_num in range(doc.page_count):
            page = doc.load_page(page_num)
            text += page.get_text("text")

        # Since no outline exists, treat the whole text as a single chunk initially
        list_of_documents = []

        if self.__len(text) > self.chunk_size:
            # Use fallback splitter for long text
            for t in self.fall_back_splitter.split_text(text):
                list_of_documents.append(Document(page_content=t, metadata={"heading": "Full Text"}))
        else:
            list_of_documents.append(Document(page_content=text, metadata={"heading": "Full Text"}))

        return list_of_documents
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

model_name = "mixedbread-ai/mxbai-embed-large-v1"
model_kwargs = {'trust_remote_code':True}
encode_kwargs = {'normalize_embeddings': False}
hf = HuggingFaceEmbeddings(
    model_name=model_name,
    encode_kwargs=encode_kwargs,
    model_kwargs=model_kwargs
)
embedding_function = hf
```

```python
from langchain.chains import RetrievalQA
from langchain.chains.combine_documents.stuff import StuffDocumentsChain
from langchain.vectorstores import Chroma
from langchain_core.output_parsers import StrOutputParser

# --- Step 1: Split PDF into smaller chunks ---
rsps = PersianPDFSplitter(chunk_size=1000, chunk_overlap=100)
documents = rsps.create_documents("/content/Drugs.pdf")
print(f"Number of documents: {len(documents)}")

# --- Step 2: Create vector store & retriever ---
vectorstore = Chroma.from_documents(
    documents=documents,
    embedding=hf,
    collection_name="drugs_pdf"
)
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})  # only top 3 chunks

# --- Step 3: Build safe RAG chain ---
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever,
    chain_type="stuff",  # safely combines retrieved chunks
    return_source_documents=False
)

# --- Step 4: Persian questions ---
persian_questions = [
    "دیکلوفناک چیست و چه کاربردی دارد؟",
    "شیاف دیکلوفناک برای چه مواردی استفاده می‌شود؟",
    "نحوه عملکرد دیکلوفناک چگونه است؟",
    "دوز روزانه مجاز دیکلوفناک برای بزرگسالان چقدر است؟",
    "عوارض جانبی رایج دیکلوفناک کدامند؟"
]

# --- Step 5: Run the questions ---
for question in persian_questions:
    print(f"Question: {question}")
    try:
        answer = qa_chain.run(question)
        print(f"Answer: {answer}\n")
    except Exception as e:
        print(f"Error: {e}\n")
```

```python
from langchain.retrievers.multi_vector import MultiVectorRetriever
from langchain.storage import InMemoryByteStore
from langchain_chroma import Chroma
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
from langchain.text_splitter import RecursiveCharacterTextSplitter
import uuid

# Assuming documents are loaded from RecursiveStructuralPDFSplitter
rsps = PersianPDFSplitter()
documents = rsps.create_documents("/content/Drugs.pdf")
print(f"Number of documents: {len(documents)}")

# Initialize vector store and byte store
vectorstore = Chroma(
    collection_name="full_documents",
    embedding_function=hf
)
store = InMemoryByteStore()
id_key = "doc_id"
retriever = MultiVectorRetriever(
    vectorstore=vectorstore,
    byte_store=store,
    id_key=id_key,
)

# Generate unique IDs for documents
doc_ids = [str(uuid.uuid4()) for _ in documents]

# Split documents into smaller chunks
child_text_splitter = RecursiveCharacterTextSplitter(chunk_size=400, chunk_overlap=50)
sub_docs = []
for i, doc in enumerate(documents):
    _id = doc_ids[i]
    _sub_docs = child_text_splitter.split_documents([doc])
    for _doc in _sub_docs:
        _doc.metadata[id_key] = _id
    sub_docs.extend(_sub_docs)

# Add documents to vector store and docstore
retriever.vectorstore.add_documents(sub_docs)
retriever.docstore.mset(list(zip(doc_ids, documents)))

# Define RAG chain
rag_chain = (
    {"context": retriever | (lambda docs: docs[:3]), "question": RunnablePassthrough()}  # Limit to 3 chunks
    | prompt
    | llm
    | StrOutputParser()
)

# Test questions based on Drugs.pdf content (Persian)
questions = [
    "عوارض جانبی رایج دیکلوفناک کدامند؟",
    "دوز روزانه مجاز دیکلوفناک برای بزرگسالان چقدر است؟",
    "نحوه عملکرد دیکلوفناک چگونه است؟"
]

# Execute and print results with error handling
for question in questions:
    print(f"Question: {question}")
    try:
        answer = rag_chain.invoke(question)
        print(f"Answer: {answer}")
    except AssertionError as e:
        print(f"Error: {e}. Check model configuration or reduce context size.")
    except ValueError as ve:
        print(f"Error: {ve}. Ensure max_model_len is sufficient or reduce chunks.")
    print()
```

```python
import pandas as pd

# List of Persian questions based on Drugs.pdf content
questions = [
    "دیکلوفناک چیست و چه کاربردی دارد؟",
    "شیاف دیکلوفناک برای چه مواردی استفاده می‌شود؟",
    "آمپول دیکلوفناک چه کاربردی دارد؟",
    "دیکلوفناک چگونه عمل می‌کند؟",
    "دوز روزانه قرص دیکلوفناک چقدر است؟",
    "نحوه استفاده از پماد دیکلوفناک چگونه است؟",
    "عوارض جانبی رایج دیکلوفناک کدامند؟",
    "عوارض جدی دیکلوفناک چیست؟",
    "چه کسانی نباید دیکلوفناک مصرف کنند؟",
    "کلیندامایسین برای چه عفونت‌هایی تجویز می‌شود؟"
]

# Initialize empty list to store answers
answers = []

# Process each question with rag_chain and store answers
for q in questions:
    try:
        answer = rag_chain.invoke(q)
        answers.append(answer)
    except Exception as e:
        answers.append(f"Error: {e}")
        print(f"Error processing '{q}': {e}")

# Create a DataFrame to store questions and answers
df = pd.DataFrame({
    "Question": questions,
    "Answer": answers
})

# Display the DataFrame
print(df)
```
